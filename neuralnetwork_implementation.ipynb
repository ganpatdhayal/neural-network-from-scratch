```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import math

# Set seed
np.random.seed(42)

# Generate a classification dataset
x, y = make_classification(
    n_samples=200,
    n_features=2,          # Only 2 useful features for visualization
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    class_sep=0.8,         # Low separation makes it harder
    flip_y=0.1,            # Add noise (10% label flipping)
    random_state=42
)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x[y == 0][:, 0], x[y == 0][:, 1], color="red", label="Class 0", alpha=0.6)
plt.scatter(x[y == 1][:, 0], x[y == 1][:, 1], color="blue", label="Class 1", alpha=0.6)
plt.title("Dummy Binary Classification Data (for Logistic Regression)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape:", x.shape)
print("Target shape:", y.shape)
```

```python
y= y.reshape(-1,1)
x_mean = np.mean(x)
x_std = np.std(x)
x = (x-x_mean)/x_std
x1 = x[:,0].reshape(-1,1)
x2 = x[:,1].reshape(-1,1)
def preprocess_input(x):
    x1 = x[:, 0].reshape(-1, 1)
    x2 = x[:, 1].reshape(-1, 1)
    return np.hstack((x1, x2))
x = np.hstack((x1,x2))
x_preprocessed = preprocess_input(x)
m = x_preprocessed.shape[0]

def sigmoid(z):
     return 1/(1+np.exp(-z))


def d_sigmoid(z):
    s = sigmoid(z)
    return s*(1-s)

def ReLu(z):
    return np.maximum(0,z)

def d_ReLu(z):
    return (z > 0).astype(float)

def dense(A_in,W,B,activation):
    Z = A_in@W +B
    if(activation == 'ReLu'):
        A_out = ReLu(Z)
    elif(activation == 'sigmoid'):
        A_out = sigmoid(Z)
    return Z,A_out
layers = [2,50,50,50,50,1]
activation_fn = ["ReLu","ReLu","sigmoid","sigmoid","sigmoid"]
lr = 0.04
np.random.seed(0)
w = [np.random.randn(layers[i],layers[i+1])for i in range(len(layers)-1)]
b = [np.zeros((1,layers[i+1]))for i in range(len(layers)-1)]
def sequential(x):
    A = x
    z_values = []
    a_values = [x]
    for i in range(len(w)):
        Z, A = dense(A, w[i], b[i], activation_fn[i])
        z_values.append(Z)
        
        a_values.append(A)
    return z_values, a_values

def backward(y,z_values,a_values,lr):
    global w,b
    m = y.shape[0]
    L  = len(w)
    dz = a_values[-1]-y

    for i in reversed(range(L)):
        A_prev = a_values[i]
        dw = (A_prev.T @ dz) / m
        db = np.sum(dz, axis=0, keepdims=True) / m
        w[i] -= lr * dw
        b[i] -= lr * db

        if i > 0:
            da = dz @ w[i].T
            if activation_fn[i-1] == 'ReLu':
                dz = da * d_ReLu(z_values[i-1])
            else:
                dz = da * d_sigmoid(z_values[i-1])




iteration = 50000
cost = []
for i in range(iteration):
    x_preprocessed = preprocess_input(x)
    z_values,a_values = sequential(x_preprocessed)
    
    a_final = a_values[-1]
    epsilon = 1e-8
    cst = -np.mean(y*np.log(a_final + epsilon) + (1 - y)*np.log(1 - a_final + epsilon))
    cost.append(cst)

    backward(y, z_values, a_values, lr)


def predict(x):
    x = preprocess_input(x)
    _,a_values = sequential(x)
    return a_values[-1]
preds = predict(x)
binary_predictions = (preds >= 0.5).astype(int)
accuracy = np.mean(binary_predictions == y)
print(f"Accuracy:  {accuracy*100:.2f}%")
plt.plot(cost)
print(cost[-1])
print("Min prediction:", np.min(preds))
print("Max prediction:", np.max(preds))
```

```python
# Plot decision boundary
h = 0.01  # step size in the mesh

# Original x1 and x2
x1_min, x1_max = x[:, 0].min() - 1, x[:, 0].max() + 1
x2_min, x2_max = x[:, 1].min() - 1, x[:, 1].max() + 1

xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h),
                       np.arange(x2_min, x2_max, h))

# Create input grid
grid_input = np.c_[xx1.ravel(), xx2.ravel()]
grid_input_processed = preprocess_input(grid_input)

# Predict for the entire grid
Z = predict(grid_input)
Z = Z.reshape(xx1.shape)

# Plot the contour and training points
plt.contourf(xx1, xx2, Z, levels=[0, 0.5, 1], cmap="coolwarm", alpha=0.6)
plt.scatter(x[:, 0], x[:, 1], c=y.ravel(), edgecolors='k', cmap="coolwarm")
plt.title("Decision Boundary")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
